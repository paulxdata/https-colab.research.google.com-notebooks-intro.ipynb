```python



import pandas as pd

# Create student data
data = {
    "Name": ["damain", "Mary", "Peter", "Grace", "Paul",
             "john", "David", "caleb", "James", "danie"],
    "Math": [90, 77, 55, 66, 88, 75, 88, 54, 67, 90],
    "English": [65, 90, 50, 92, 55, 90, 85, 60, 70, 93],
    "Science": [80, 88, 12, 98, 58, 78, 90, 65, 72, 99]
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("students_week6.csv", index=False)

print("CSV file created successfully!")


df = pd.read_csv("students_week6.csv")
df.head()

subject_averages = df[["Math", "English", "Science"]].mean()
print("Subject Averages:\n", subject_averages)

df["Average"] = df[["Math", "English", "Science"]].mean(axis=1)
print("\nStudent Averages:\n", df[["Name", "Average"]])

top_student = df.loc[df["Average"].idxmax()]
print("\nTop Performing Student:\n", top_student["Name"], "-", round(top_student["Average"], 2))

lowest_student = df.loc[df["Average"].idxmin()]
print("Lowest Performing Student:\n", lowest_student["Name"], "-", round(lowest_student["Average"], 2))

df.to_csv("students_week6_updated.csv", index=False)
print("\nUpdated CSV exported successfully!")






print('paul amarachukwu udeh')
print('202550749955HA')
```
