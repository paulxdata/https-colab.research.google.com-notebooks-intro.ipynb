```python
#create a list of 5 students scores
scores = [75, 88, 92, 64, 81]
print(scores)

#compute the average score
average = sum(scores) / len(scores)
print('average score:', average)

#find the highest score
highest = max(scores)
print('higest score:', highest)


#craete a dictionary with 5 students and scores
students = {'damian': 75,
            'paul': 88,
            'calebe': 92,
            'jerry': 64,
            'emma': 81,
            }
print(students)

#find the highest scoring student
highest_student = max(students, key=students.get)
print('highest scoring students:', highest_student)
print('highest score:', students[highest_student])

#create a tuple of subjects
subjects = ('maths', 'english', 'biology', 'physics', 'chemistry',)
print('subjects:', subjects)

#create a set of unique scores
unique_scores = set(students.values())
print('unique scores:', unique_scores)
```
